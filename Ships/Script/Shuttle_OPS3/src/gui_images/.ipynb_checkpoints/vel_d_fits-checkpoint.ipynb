```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

import sklearn
print(sklearn.__version__)
```

```python
# Load the CSV file
data = pd.read_csv('vel-drag-fit1.csv')

# Pivot the DataFrame
data = data.melt(id_vars=['y'], var_name='d', value_name='x').astype('float64').dropna(axis= 0, how='any')

# Sort the DataFrame by 'foo' column
data.sort_values(['x','d'], ignore_index=True, inplace=True)

data.head(20)
```

```python
# Reshape the DataFrame into a pivot table
pivot_table = data.pivot(index='d', columns='y', values='x')



# Convert the data to a 2D numpy array
X = pivot_table.values


# Create the X and Y coordinate grids
Y, D = np.meshgrid(pivot_table.columns, pivot_table.index)


# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface, ignoring NaN values
ax.plot_surface(X, Y, D, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('drag')
ax.set_title('traj')



ax.view_init(90,270)

# Show the plot
plt.show()
```

```python
# Flatten the X, Y, and Z arrays
X_flat = X.flatten()
Y_flat = Y.flatten()
D_flat = D.flatten()


degree = 3  # Set the degree of the polynomial
poly_features = PolynomialFeatures(degree=degree, interaction_only=False)
YD_poly = poly_features.fit_transform(np.column_stack((Y_flat, D_flat)))

# Perform polynomial regression
model = LinearRegression()
model.fit(YD_poly, X_flat)

# Generate predictions for the grid points
X_pred = model.predict(YD_poly)

# Reshape the predicted values back to a 2D array
X_pred = X_pred.reshape(X.shape)

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the original data points
ax.scatter(X_flat, Y_flat, D_flat, c='red', label='Original Data')

# Plot the predicted surface
ax.plot_surface(X_pred, Y, D, cmap='viridis', alpha=0.7, label='Polynomial Fit')

# Set labels and title
# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('drag')
ax.set_title('x-vs-drag')
ax.set_title('Polynomial Fit of x')

# Add a legend


ax.invert_xaxis()

ax.view_init(10,20)


# Show the plot
plt.show()
```

```python
# Retrieve the coefficients from the trained model
coef = model.coef_
intercept = model.intercept_

# Define the equation for the prediction
equation = f"x = {intercept}"

for i, c in enumerate(coef[1:], start=1):
    power = poly_features.powers_[i]
    
    equation += f" + {c} * y^{power[0]} * drag^{power[1]}"

print("Equation for the model prediction:")
print(equation)
print("Equation in python form:")
print(equation.replace('^','**'))
```
